1)Create AWS account


2)Install WinSCP, Putty



3)Generate Putty private key(.ppk) with reference to : 
https://asf.alaska.edu/how-to/data-recipes/connect-to-your-ec2-instance-using-putty-v1-1/



4)Create EC2 instances as required

5)Open the AWS Instances window in the EC2 Management Console.
The Public DNS of your EC2 instance displayed in this window will be used in the Transferring Files Using WinSCP section and the Transferring Files Using PuTTY Secure Copy (PSCP) section.

6) Transferring Files Using WinSCP

WinSCP Login window.
Click on New Site.
In the Host name box, enter the Public DNS displayed in your EC2 Management Console Instances window.
In the User name box, type “ubuntu.”
Click on the Advanced button.
Advanced Site Settings window
Under SSH (1), click on Authentication (2)
Under Private key file (3), click on the … button and navigate to the folder where your PuTTY Private Key (.ppk) file is located.
Select the .ppk file.
Click on the OK button to close the Advanced Site Settings window.
Click the Login button
The first time you connect to your instance, you will be asked about connecting to an unknown server.
Click the Yes button to continue.

7)Open a Windows “Command Prompt” window
When using PSCP to transfer files between your computer and your EC2 instance, your “PuTTY Private Key (.ppk)” file must be in the folder you are working in, or you must provide a path in the PSCP command to the folder the file is located. For example, if “AWSkey.ppk” file is stored in the folder named “keys,” include this path in the PSCP command:

C:\Users\username\keys\AWSkey.ppk(for example)

```
#At the Windows command prompt type (notice where spaces are placed):

C:\> pscp -i yourkey.ppk yourfilename ubuntu@public_DNS:/home/ubuntu/
```

***8) File should be transferred to the instance you desire!***
